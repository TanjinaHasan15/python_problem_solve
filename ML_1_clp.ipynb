```python
dataset = pd.read_csv('house_price_regression_dataset.csv')
dataset.head()
```

```python
dataset.plot(x='Square_Footage',y='House_Price',style='1')
plt.title('House price vs Square footage')
plt.xlabel('Square_Footage')
plt.ylabel('House_Price')
plt.show()
```
