```
import pandas as pd
import numpy as np
from sklearn import datasets
from sklearn.cluster import KMeans
import matplotlib.pyplot as plt
import matplotlib.patches as mpatches
import sklearn.metrics as sm
%matplotlib inline
```

```
iris = datasets.load_iris ()   #loading dataset
```

```
print (iris.data [0:5])    #Prining the first 5 data
```

```
print (iris.target_names)   #prining the target data names
```

```
print (iris.target)
```

```
# assigning column labels SL, SW, PL, PW for SL: Sepal Length ,
# SW: Sepal Width , PL: Petal Length and PW: Petal Width.
x = pd.DataFrame(iris.data , columns=['SL','SW','PL','PW'])   #two-dimensional data structure
y = pd.DataFrame(iris.target , columns=['Target'])
```

```
x.head()
```

```
y.head()   # in data set there is no target colum
```

```
plt.figure(figsize =(12 ,3))
colors = np.array(['red', 'green', 'blue'])
iris_targets_legend = np.array(iris.target_names)
red_patch = mpatches.Patch(color='red', label='Setosa')
green_patch = mpatches.Patch(color='green',label='Versicolor')
blue_patch = mpatches.Patch(color='blue',label='Versicolor')
plt.subplot(1, 2, 1)
plt.scatter(x.SL , x.PL , c=colors[y.Target])
plt.xlabel('Sepal Length')
plt.ylabel('Petal Length')
plt.legend(handles=[red_patch, green_patch, blue_patch])
plt.subplot (1,2,2)
plt.scatter(x['PL'], x['PW'], c= colors[y['Target']])
plt.title('Petal Length vs Petal Width')
plt.legend(handles=[red_patch , green_patch , blue_patch ])
```

```
iris_k_mean_model = KMeans(n_clusters =3) # we set K = 3
iris_k_mean_model.fit(x) # trained using the entire dataset
```

```
print(iris_k_mean_model.predict(x))   #to gain an insight about the KMeans algorithm.
```

```
print (iris_k_mean_model.cluster_centers_)   #identify which cluster center belongs which class of iris flower
```

```
plt.figure(figsize =(12 ,3))

colors = np.array(['red', 'green', 'blue'])

predictedY=np.choose(iris_k_mean_model.labels_ ,[1,0,2]).astype(np.int64)

plt.subplot(1, 2, 1)
plt.scatter(x['PL'], x['PW'], c=colors[y['Target']])
plt.title('Before classification')
plt.legend(handles=[red_patch , green_patch , blue_patch ])

plt.subplot(1, 2, 2)
plt.scatter(x['PL'], x['PW'], c=colors[predictedY ])
plt.title("Model's classification")
plt.legend(handles=[red_patch , green_patch , blue_patch ])
```

```
sm.accuracy_score(predictedY , y['Target'])
```

```
sm.confusion_matrix(predictedY , y['Target'])
```
